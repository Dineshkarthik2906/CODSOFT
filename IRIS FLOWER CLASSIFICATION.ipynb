```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Documents/iris.csv')
```

```python
df.columns = ["sepal_length","sepal_width","petal_length","petal_width","target"]
```

```python
df.target.replace({"Iris-setosa":"setosa","Iris-versicolor":"versicolor","Iris-virginica":"virginica"},inplace=True)
```

```python
#feature_scaling
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
```

```python
X = df.drop(['target'],axis=1)
```

```python
y = df['target']
```

```python
scaler = ss.fit(X)
```

```python
X_stdscaler = scaler.transform(X)
```

```python
X_stdscaler
```

```python
#label_encoder
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
y = le.fit_transform(y)
```

```python
y
```

```python
#splitting
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_stdscaler,y,test_size=0.3,random_state=42) 
```

```python
#selection_0f_model
from sklearn.svm import SVC
```

```python
svc = SVC(kernel="rbf")
svc.fit(X_train,y_train)
```

```python
y_pred = svc.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_pred,y_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train,y_train)
```

```python
y_pred = knn.predict(X_test)
```

```python
accuracy_score(y_pred,y_test)
```
